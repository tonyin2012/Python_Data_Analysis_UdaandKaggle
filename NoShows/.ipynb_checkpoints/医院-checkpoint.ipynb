```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# 查看数据

```python
df = pd.read_csv("noshowappointments-kagglev2-may-2016.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
mapping = {'No': 0,'Yes' : 1}
df['No-show'] = df['No-show'].map(mapping)
df.head()
```

```python
(df['Age']<=0).sum()
```

```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['ScheduledDate'] = pd.to_datetime(df['ScheduledDay'].dt.strftime('%Y-%m-%d'))
df['ScheduledHour'] = pd.to_datetime(df['ScheduledDay'].dt.strftime('%H:%M:%S'))
```

```python
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
```

```python
for x in range(len(df)):
    if df['AppointmentDay'][x] < df['ScheduledDate'][x]:
        df['AppointmentDay'][x] =df['ScheduledDate'][x]
```

```python
df['time'] = df['AppointmentDay']- df['ScheduledDate']
df['time'].describe()
```

```python
df[df['time']<pd.Timedelta(-1,'D')]
```

```python
df['show'] = pd.get_dummies(df['No-show'])['No']
df['show']
```

```python
No_Show_Age = df.query('show== 0')['Age']
```

```python
fig, axes = plt.subplots(1, 1, figsize=(20, 10))
plt.hist(df.query('show== 0')['Age'])
plt.xlim((0,120))
```

```python
df['Age'].describe()
```

```python
bin_edge = [0,18,37,55,115]
bin_names = [ 'Teens', 'Youth', 'MiddleAge', 'Old']
df['Age_Level'] = pd.cut(df['Age'], bin_edge, labels=bin_names)
df.head(1)
```

```python
agedf = (df.groupby('Age_Level')['show'].value_counts()/df.groupby('Age_Level')['PatientId'].count()).reset_index(name = 'proportion')
agedf
```

```python
import seaborn as sns
sns.set_style ('dark')
#柱状图图每个坐标点总宽度
total_width= 0.8
#每个坐标点共有柱状图数量
n = 2
#坐标点总数
size = 4
#每个柱子宽度
width = total_width / n
#调整坐标轴位置
x = np.arange(size) - (total_width - width) / 2
# 不出席的值
a = agedf.query('[0] in show ')["proportion"].values
# 出席的值
b = agedf.query('[1] in show ')["proportion"].values
#不出席的病患里年龄平均值
aver_noshow = (df.groupby('show')['Age'].count()/df['Age'].count())[0]
#出席的病患里年龄平均值
aver_show = (df.groupby('show')['Age'].count()/df['Age'].count())[1]
#画板大小
plt.figure(figsize=(8,8), dpi=80)
#不出席的柱状图
plt.bar(x,a,width=width, label='no show',color ='pink')
#出席的柱状图
plt.bar(x + width, b, width=width, label='show',color ='orange')
#不出席的平均值的折线图
plt.plot(np.arange(6)-1,[aver_noshow] * 6,ls='--', marker = 'D',label='no show mean',color ='grey')
#出席的平均值的折线图
plt.plot(np.arange(6)-1,[aver_show] * 6,ls='--', marker = '*',label='show mean',color ='SeaGreen')
#修改x轴坐标文本
plt.xticks(x + width/2,agedf['Age_Level'].unique())
#确定x和y轴坐标范围
plt.xlim((-1,4))
plt.ylim((0,1))
#创建图例
plt.legend()
#标题
plt.title('proportion of age absent',fontsize = 18)
plt.xlabel('Age',fontsize = 18)
plt.ylabel('proportion',fontsize = 18)
plt.show()
```
